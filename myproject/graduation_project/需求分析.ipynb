什么是共享经济？

今日人们理解的共享经济已经上升为一种商业模式，即拥有闲置资源的某个机构或者个人，以有偿的方式将资源使用权暂时让渡给他人，并获取一定报酬。

深入挖掘，不难发现，共享经济模式的主要特征有以下三点：

第一，借助网络搭建信息平台。如此可以打破供需双方的信息不对称，及时高效地连接闲置物品或服务的所有者与需求者，在大大降低搜寻成本的同时，还能让达成交易的可能性得到提升。

第二，以闲置资源使用权的暂时性转移为本质。共享经济倡导的是所有权与使用权的分离，物品或服务的需求者通过共享平台暂时性地从所有者那里获得使用权，以相对较低的成本完成使用目标后，再将物品归还给所有者。

第三，重视物品的重复交易和高效利用。通过频繁易手所有者的闲置资源，让其他社会成员反复使用，以这种“网络串联”的方式将全社会范围内的闲置资源重新加以配置、整合与优化，从而提升现有物品的使用效率，并实现个体的福利提升与社会整体的绿色可持续发展，这也是共享经济的精髓所在。

# UI相关

把ui的功能详细描述

```
区块链数据服务子系统
【用户/设备资产，余额记录，租用记录】
1）创建用户/设备资产
在区块链数据服务中，用户和设备都会与现实资产进行绑定，实现设备的身份鉴权。
2）余额记录
在租赁业务中，用户的余额即数字货币，会因为设备租赁操作和充值等操作变更，资产拥有者和使用者的余额都会被安全的更新。
3）租用记录
每台设备的租用都会被按精确的时间记录，已达到精确计费的目的。可以对资产拥有者和使用者进行交易结算。

物联网通信服务子系统
【（mqtt） 代理服务器，客户端】
4）mqtt消息代理服务器
为mqtt客户端提供发布订阅模式的消息队列服务，用于整个系统间的通信服务。
5）mqtt客户端
用于各子系统连接mqtt代理服务器，进行通信的客户端，连接设备，共享服务，区块链数据服务。

前端服务子系统
【用户注册，设备注册，设置共享；
用户登陆，开始租用，停止租用；
查看设备信息；租用历史；
充值，兑换；
后台管理（用户管理，设备管理）】
6）用户注册
当新用户使用设备前，需要注册，系统会为用户提供用于区块链数据服务的非对称密钥，用户的密钥钥需要保存在用户本地，作为以后每次操作的身份凭证。
7）设备注册
当用户要共享自己的车位时，提供车位的相关描述，注册成为共享资源。
8）共享设置
车位拥有者可以设置车位的共享时段和租金，修改共享状态。
9）充值/套现
当资产拥有者和使用者需要充值或套现数字货币时，系统维护方作为交易者使用现金交易换取用户的数字货币。
10）开始/中止租用
当使用者需要开始和中止租用车位的时候，使用手机扫描设备识别码，可以查看设备的共享信息，选择开始/中止租用设备，中止使用设备后，会进行结算。
11）后台管理
后台管理员，可以管理用户和设备，进行封禁等操作。
```
